### Data Cleaning

```python
# dropping all date columns from remaining columns
# need to use the date columns for derived variables (use data dataframe)

date_cols=['date_of_last_rech_data_7','date_of_last_rech_data_6','date_of_last_rech_data_8','date_of_last_rech_data_9','date_of_last_rech_6','date_of_last_rech_7','date_of_last_rech_8','date_of_last_rech_9', 'last_date_of_month_7','last_date_of_month_8','last_date_of_month_9']
data.drop(columns=date_cols, inplace=True)
data.shape
```

```python
cols_to_drop = ['mobile_number']
for i in data.columns:
      if len(data[i].dropna().unique()) == 1:
            cols_to_drop.append(i)
```

```python
data.drop(columns=cols_to_drop, inplace=True)
```

```python

round(data.isnull().sum(axis=0)/len(data)*100,2)
```

```python
## sachet, monthly, night pack, fb user
n = data.filter(regex= 'night_',axis = 1).columns
f = data.filter(regex= 'fb_',axis = 1).columns
on = data.filter(regex= 'onnet_',axis = 1).columns
off = data.filter(regex= 'offnet_',axis = 1).columns

```

```python
# Meaningful Imputation
for i in n:
    print(i)
    m = data[i].mode()[0]
    data[i] = data[i].fillna(m) #,inplace = True)
for i in f:
    data[i] = data[i].fillna(data[i].mode()[0]) #,inplace = True)
```

```python

round(data.isnull().sum(axis=0)/len(data)*100,2)
```

```python
for i in data.columns:
    if (data[i].isnull().sum()/len(data))*100 > 70:
        data.drop(i,inplace = True, axis = 1)
data.shape
```

```python
cat_cols_ = []
```

```python

for i in on:
    cat_cols_.append(i) 

for i in off:
    cat_cols_.append(i)
cat_cols_
```

```python
cat_cols_
```

```python
for i in cat_cols_:# impute NaN with Mode since these are meaningfull Na
    data[i].fillna(data[i].mode(), inplace=True)
    
data.head()
```

```python
# use binning for creating categorical cols
cat_labels_67 = ['Low_onnet_67', 'Average_onnet_67', 'High_onnet_67']
cat_labels_6 = ['Low_onnet_6', 'Average_onnet_6', 'High_onnet_6']
cat_labels_7 = ['Low_onnet_7', 'Average_onnet_7', 'High_onnet_7']
cat_labels_8 = ['Low_onnet_8', 'Average_onnet_8', 'High_onnet_8']

df['onnet_67'] = pd.qcut(df.onnet_mou_6+df.onnet_mou_7,
                              q=[0,.33,.66,1],
                              labels=cat_labels_67)
df['onnet_6'] = pd.qcut(df.onnet_mou_6,
                              q=[0,.33,.66,1],
                              labels=cat_labels_6)
df['onnet_7'] = pd.qcut(df.onnet_mou_7,
                              q=[0,.33,.66,1],
                              labels=cat_labels_7)
df['onnet_8'] = pd.qcut(df.onnet_mou_8,
                              q=[0,.33,.66,1],
                              labels=cat_labels_8)
df.drop(columns=['onnet_mou_6','onnet_mou_7','onnet_mou_8','onnet_mou_9'], inplace=True)
```

### 3. Outlier Treatment

```python
df.head()
```

```python
#num_cols.remove('churn')
df[num_cols].describe(percentiles=[.25, .5, .75, .90, .95, .99])
```

```python

for i in enumerate(num_cols):
        Q1 = df[i[1]].quantile(0.010)
        Q3 = df[i[1]].quantile(0.99)
        IQR = Q3 - Q1
        lower = Q1- IQR*1.5
        upper = Q3+ IQR*1.5
        df = df[(df[i[1]] >= lower) & (df[i[1]] <= upper)]

print(df.shape)
```

```python
from IPython.display import Image
from io import StringIO
from sklearn.tree import export_graphviz
import pydotplus, graphviz
from sklearn.ensemble import RandomForestClassifier
```

```python
def evaluate_model(dt_classifier):
    print("Train Accuracy :", accuracy_score(y_train, dt_classifier.predict(X_train)))
    print("Train Recall :", recall_score(y_train, dt_classifier.predict(X_train)))
    print("Train Precision :", precision_score(y_train, dt_classifier.predict(X_train)))
    print("Train F1 :", f1_score(y_train, dt_classifier.predict(X_train)))
    print("Train Confusion Matrix:")
    print(confusion_matrix(y_train, dt_classifier.predict(X_train)))
    print(classification_report(y_train, dt_classifier.predict(X_train)))
    print("-"*50)
    print("Test Accuracy :", accuracy_score(y_test, dt_classifier.predict(X_test)))
    print("Test Recall :", recall_score(y_test, dt_classifier.predict(X_test)))
    print("Test Precision :", precision_score(y_test, dt_classifier.predict(X_test)))
    print("Test F1 :", f1_score(y_test, dt_classifier.predict(X_test)))
    print("Test Confusion Matrix:")
    print(confusion_matrix(y_test, dt_classifier.predict(X_test)))
    print(classification_report(y_test, dt_classifier.predict(X_test)))
    
```

```python
def get_dt_graph(dt_classifier):
    dot_data = StringIO()
    export_graphviz(dt_classifier, out_file=dot_data, filled=True,rounded=True,
                    feature_names=X.columns, 
                    class_names=['Not Churn', "Churn"])
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    return graph
```

### Observations from Model output and Recommendations
- Model without dimentionality reduction is giving better interpretability and accuracy
- Among other models with dimentionality reduction, better model suggestions are in the order ,  Logistic Regression, Decision Tree, Random Forest as we need to consider Recall 
- Non-Tree based models (Logistic Regression) with PCA are showing better results than tree based models
- Recall is the metric to be taken into account as we need to concentrate on both True Positives and False Negatives on the Churn target attribute
- With data from months 6 and 7 used for determining high value customers and data from month 9 is used to determine if a customer churns, it falls on month 8 data to take any necessary action to make a customer non churn

- Decission Tree is recommended in place of a Random forest as the recall metric is better
- In case of large dataset and model building is taking too long, PCA is recommended for dimentionality reduction to improve performance
- top deciding attributes to determine if a customer may churn are:
    - local calls (incoming and outgoing)
    - recharge amount
    - roaming calls (outgoing)
- Since the dataset is old, no interesting data found to rely on internet usage
- in case if a customer is receving low incoming calls and making low minutes of usage on outgoing calls, it can be understood that he/she is trying a new service provider with better call rates/offers. In such cases, it is recommended to advertise the customer with better offers 
- As one of the top attributes is roaming calls, any customer moving from one state to another is more likely to churn if planned to stay long in other states; if any new plans were introduced to reduce the roaming incoming and outgoing calls, customers are more likely continue to use the services
